```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DQN_lane = pd.read_csv('DQN_Controller_metrics_lane.csv')
MP_lane = pd.read_csv('max_pressure_metrics_lane.csv')

DQN_veh = pd.read_csv('DQN_Controller_metrics_veh.csv')
MP_veh = pd.read_csv('max_pressure_metrics_veh.csv')

DQN_lane.head()
```

```python
DQN_lane_step = DQN_lane.groupby('step').agg({'WaitingTime':'mean','CO2':'mean'}).reset_index()
MP_lane_step = MP_lane.groupby('step').agg({'WaitingTime':'mean','CO2':'mean'}).reset_index()

DQN_vehicle_step = DQN_veh.groupby('step').agg({'TimeLoss':'mean','AccumulatedWaitingTime':'mean'}).reset_index()
MP_vehicle_step = MP_veh.groupby('step').agg({'TimeLoss':'mean','AccumulatedWaitingTime':'mean'}).reset_index()

DQN_lane_step.head()
```

```python
fig = plt.figure()

plt.plot(DQN_lane_step['step'],DQN_lane_step['WaitingTime'], label = 'DQN')
plt.plot(MP_lane_step['step'],MP_lane_step['WaitingTime'], label = 'MP')
fig.legend()
```

```python
fig = plt.figure()

plt.plot(DQN_lane_step['step'],DQN_lane_step['CO2'], label = 'DQN')
plt.plot(MP_lane_step['step'],MP_lane_step['CO2'], label = 'MP')
fig.legend()
```

```python
fig = plt.figure()
plt.plot(DQN_vehicle_step['step'],DQN_vehicle_step['TimeLoss'], label = 'DQN')
plt.plot(MP_vehicle_step['step'],MP_vehicle_step['TimeLoss'], label = 'MP')
fig.legend()
```

```python
fig = plt.figure()
plt.plot(DQN_vehicle_step['step'],DQN_vehicle_step['AccumulatedWaitingTime'], label = 'DQN')
plt.plot(MP_vehicle_step['step'],MP_vehicle_step['AccumulatedWaitingTime'], label = 'MP')
fig.legend()
```

```python
vehicle = pd.read_csv('AAA_metrics_veh.csv')
vehicle.head()
```

```python
vehicle_step = vehicle.groupby('step').agg({'TimeLoss':'mean','AccumulatedWaitingTime':'mean'}).reset_index()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(vehicle_step['step'],vehicle_step['TimeLoss'],'-', label = 'TimeLoss')

ax2 = ax.twinx()
ax2.plot(vehicle_step['step'],vehicle_step['AccumulatedWaitingTime'],'r', label = 'AccumulatedWaitingTime')
fig.legend()

ax.set_xlabel('Time Step')
ax.set_ylabel('Time Loss')
ax2.set_ylabel('AccumulatedWaitingTime')
```
